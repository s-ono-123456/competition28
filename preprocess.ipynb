```python
train = train.reset_index().drop('index', axis=1)
train.loc[train['地区名'].isna(), '地区名'] = train['市区町村名']
```

```python
# 面積から間取りを予測する。
```

```python
# 面積の項目に2000㎡以上文字列があるため、一律2000へ置き換える。
```

```python
train.loc[train['面積（㎡）'] == '2000㎡以上', '面積（㎡）'] = '2000'
train['面積（㎡）'] = train['面積（㎡）'].astype(int)
madoriindex = train[train['間取り'].isna()].index

madori = train[train['間取り'].notna()]
madori = madori.groupby('間取り').mean()

for index in madoriindex:
    menseki = train.iloc[index,:]['面積（㎡）']
    train.iloc[index, train.columns.get_loc('間取り')] = (madori['面積（㎡）']-menseki).abs().idxmin()
```

```python
train.isnull().sum()
```

```python
# 都市計画、建ぺい率、容積率、最寄り駅：距離はいずれも市区町村コードと地区名から最頻値を取得して設定する。
# 20分近い時間がかかる・・・
# 処理方式変えたほうがいい・・・
```

```python
colnamelist = ['都市計画', '建ぺい率（％）', '容積率（％）', '最寄駅：名称', '最寄駅：距離（分）', '建物の構造']
for colname in colnamelist:
    print(colname)
    toshi = train[train[colname].notna()]
    toshi = toshi.groupby(['市区町村コード', '地区名'])[colname].apply(lambda x: x.mode())
    toshi = toshi.reset_index().drop('level_2', axis=1)
    city = train[train[colname].notna()]
    city = city.groupby(['市区町村コード'])[colname].apply(lambda x: x.mode())
    city = city.reset_index().drop('level_1', axis=1)
    toshiindex = train[train[colname].isna()].index
    
    for index in toshiindex:
        citycode = train.iloc[index,train.columns.get_loc('市区町村コード')]
        cityname = train.iloc[index,train.columns.get_loc('地区名')]
        data = toshi[((toshi['市区町村コード']== citycode) & (toshi['地区名'] == cityname))]
        if len(data) == 0:
            if len(city[city['市区町村コード']== citycode][colname]) == 0:
                train.iloc[index, train.columns.get_loc(colname)] = train[colname].mode()[0]
            else:
                train.iloc[index, train.columns.get_loc(colname)] = city[city['市区町村コード']== citycode][colname].iloc[-1]
        else:
            train.iloc[index, train.columns.get_loc(colname)] = toshi[((toshi['市区町村コード']== citycode) & (toshi['地区名'] == cityname))][colname].iloc[-1]
```
